```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, DateType
from pyspark.sql import Row
from datetime import datetime

# Define schema
empleados_schema = StructType([
    StructField("EmpleadoID",   StringType(),  False),
    StructField("Nombre",       StringType(),  False),
    StructField("Apellido",     StringType(),  False),
    StructField("Departamento", StringType(),  True),
    StructField("Puesto",       StringType(),  False),
    StructField("FechaAlta",    DateType(),    False),
    StructField("Salario",      DoubleType(),  False),
    StructField("Activo",       StringType(),  False),
])

# Sample data
empleados_data = [
    ("EMP001", "Ana",    "García",    "Tecnología",   "Data Engineer",    datetime(2021, 3, 15), 42000.00, "S"),
    ("EMP002", "Carlos", "López",     "Finanzas",     "Analista",         datetime(2020, 7, 1),  38000.00, "S"),
    ("EMP003", "María",  "Fernández", "Tecnología",   "Data Scientist",   datetime(2022, 1, 10), 47000.00, "S"),
    ("EMP004", "Luis",   "Martínez",  "RRHH",         "Técnico RRHH",     datetime(2019, 5, 20), 35000.00, "S"),
    ("EMP005", "Elena",  "Sánchez",   None,           "Consultor",        datetime(2023, 9, 5),  51000.00, "N"),
]

# Create DataFrame
df_empleados = spark.createDataFrame(empleados_data, schema=empleados_schema)

# Show results
df_empleados.show(truncate=False)
df_empleados.printSchema()
```

```python
# Data Transformation
from pyspark.sql.functions import year, month, col, expr, upper, datediff, current_date, round

empleados_transformed = (
    df_empleados.withColumn("AnioAlta",        year(col("FechaAlta")))
                .withColumn("MesAlta",         month(col("FechaAlta")))
                .withColumn("DiasEnEmpresa",   datediff(current_date(), col("FechaAlta")))
                .withColumn("AniosEnEmpresa",  round(col("DiasEnEmpresa") / 365, 1))
                .withColumn("SalarioAnual",    expr("Salario * 12"))
                .withColumn("Departamento",    upper(col("Departamento")))
)

# Show results
empleados_transformed.show(truncate=False)
```
